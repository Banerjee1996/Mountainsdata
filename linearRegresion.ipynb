```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Mountains.csv')
df.head(5)
```

```python
plt.figure(figsize=(24,12))
plt.scatter(df['Height (m)'],df['Prominence (m)'])
```

```python
sns.distplot(df['Ascents bef. 2004'],bins=100)
```

```python
sns.jointplot(x='Ascents bef. 2004',y='Failed attempts bef. 2004',data=df,kind='reg')
```

```python
df['Range'].value_counts()
```

```python
plt.figure(figsize=(24,12))
sns.countplot(df['Range'],data=df,orient='v')
```

```python
compare=df[(df['Range']=='Mahalangur Himalaya') | (df['Range']=='Hispar Karakoram') | (df['Range']=='Baltoro Karakoram')]
compare.head()
```

```python
plt.figure(figsize=(24,12))
sns.boxplot(x=compare['Ascents bef. 2004'],y=compare['Height (m)'],hue=compare['Range'])
```
